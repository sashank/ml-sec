```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,8)
```

```python
data = np.genfromtxt('datasets/diurnal-series/diurnal-example.log', delimiter=',', dtype=int)
```

```python
df = pd.DataFrame(data)
```

```python
x = x = np.arange(len(data))
y = data
m, b = np.polyfit(x, y, 1)
plt.plot(x, y, linewidth=5, color='k', label='data series')
plt.plot(x, m*x + b, '.', linewidth=3, color='k', label='regression line')
plt.xlim(5,160)
plt.ylim(40000,200000)
plt.legend(loc='upper left', prop={'size':30}, numpoints=5)
```
